```python
tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2")
tokenizer.pad_token_id = tokenizer.eos_token_id
tokenizer.padding_side = "left"
sft_model = AutoModelForCausalLM.from_pretrained("qwenzoo/utn-llm-assign2-gpt2-SFT")
sft_model.generation_config.pad_token_id = tokenizer.pad_token_id
dpo_model = AutoModelForCausalLM.from_pretrained("qwenzoo/utn-llm-assign2-gpt2-DPO")
dpo_model.generation_config.pad_token_id = tokenizer.pad_token_id
```

```python
benchmark_performance(sft_model, tokenizer, verbose=True)
```

```python
benchmark_performance(dpo_model, tokenizer, verbose=True)
```
